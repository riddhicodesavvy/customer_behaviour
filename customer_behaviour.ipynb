```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df.info
df.describe
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ' , '_')
df.columns = df.columns.str.replace('purchase_amount_(usd)' , 'purchase_amount')
```

```python
#create new column for age_group
#We divided ages into four quantile-based groups (`Young Adult`, `Adult`, `Middle-aged`, `Senior`) using `pd.qcut`.

labels =['Young Adult' , 'Adult' , 'Middle-aged' , 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
df.head(10)
```

```python
#create column purchase_frequency_days , get data from the column frequncy_purchases and map it accordingly for easier analysis

frequency_mapping = {
    "Fortnightly" : 14 ,
    "Weekly":7,
    "Monthly" : 30, 
    "Quarterly" : 90,
    "Bi-Weekly" : 14,
    "Annually" : 365 ,
    "Every 3 Months" : 90
    
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
#promo_code_applied column is redundant hence we dont need it 
df=df.drop('promo_code_used',axis=1)
```
